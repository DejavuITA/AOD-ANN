```python
# plotting libraries
import matplotlib
import matplotlib.pyplot as plt 
# numpy (math) libary
import numpy as np

from scipy import interpolate

path  = '../../data/2017_12_19/'
path1 = '../../data/2017_12_20/'
```

```python
VOA = np.loadtxt(path+'1100-VOA_calibration.tsv')
VOA.shape

m2a = np.loadtxt(path1+'1610-m2_andata.tsv')
m2r = np.loadtxt(path1+'1621-m2_discesa.tsv')
m2a.shape

m3a = np.loadtxt(path1+'1629-m3_andata.tsv')
m3r = np.loadtxt(path1+'1631-m3_discesa.tsv')

m4a = np.loadtxt(path1+'1633-m4_salita.tsv')
m4r = np.loadtxt(path1+'1632-m4_discesa.tsv')

m5a = np.loadtxt(path1+'1658-m5_salita.tsv')
m5r = np.loadtxt(path1+'1659-m5_discesa.tsv')

m6a = np.loadtxt(path1+'1701-m6_salita.tsv')
m6r = np.loadtxt(path1+'1702-m6_discesa.tsv')
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)


ax1.plot(m2a[:,0],m2a[:,1], label='1552.30nm')
ax1.plot(m2r[:,0],m2r[:,1], label='1552.30nm', ls='--')

ax1.plot(m3a[:,0],m3a[:,1], label='1552.37nm')
ax1.plot(m3r[:,0],m3r[:,1], label='1552.37nm', ls='--')

ax1.plot(m4a[:,0],m4a[:,1], label='1552.18nm')
ax1.plot(m4r[:,0],m4r[:,1], label='1552.18nm', ls='--')

ax1.plot(m5a[:,0],m5a[:,1], label='1552.24nm')
ax1.plot(m5r[:,0],m5r[:,1], label='1552.24nm', ls='--')

ax1.plot(m6a[:,0],m6a[:,1], label='1552.31nm')
ax1.plot(m6r[:,0],m6r[:,1], label='1552.31nm', ls='--')

legend = ax1.legend(loc=2, fontsize=16)

plt.show()
plt.close()
```

```python
# normalize intensity in the range 0% - 100%
nVOA = VOA
nVOA[:,1] = ( nVOA[:,1]-min(nVOA[:,1]) )/( max(nVOA[:,1])-min(nVOA[:,1]) )
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)

ax1.plot( nVOA[:,1], nVOA[:,0])

plt.show()
plt.close()
```

```python
pf2 = np.polyfit(nVOA[:,1], nVOA[:,0], 2)
p2  = np.poly1d(pf2)

pf3 = np.polyfit(nVOA[:,1], nVOA[:,0], 3)
p3  = np.poly1d(pf3)

pf4 = np.polyfit(nVOA[:,1], nVOA[:,0], 4)
p4  = np.poly1d(pf4)

pf5 = np.polyfit(nVOA[:,1], nVOA[:,0], 5)
p5  = np.poly1d(pf5)

pf6 = np.polyfit(nVOA[:,1], nVOA[:,0], 6)
p6  = np.poly1d(pf6)

pf7 = np.polyfit(nVOA[:,1], nVOA[:,0], 7)
p7  = np.poly1d(pf7)
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)

xp = np.linspace(0, 1, 101)
ax1.plot(nVOA[::-1,1], nVOA[::-1,0], '.')
#ax1.plot(xp, p2(xp), '-')
ax1.plot(xp, p3(xp), '--')
#ax1.plot(xp, p4(xp), '--')
#ax1.plot(xp, p5(xp), '--')
#ax1.plot(xp, p6(xp), '--')
ax1.plot(xp, p7(xp), '--')

plt.show()
plt.close()
```

```python
tck = interpolate.splrep(nVOA[::-1,1], nVOA[::-1,0], s=0)
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)

xp = np.linspace(0, 1, 101)
ax1.plot(nVOA[:,1], nVOA[:,0], '.')
ax1.plot(xp, p7(xp), '--')
ynew = interpolate.splev(xp, tck, der=0)

plt.show()
plt.close()
```
