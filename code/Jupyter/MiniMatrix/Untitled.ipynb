```python
# plotting libraries
import matplotlib
import matplotlib.pyplot as plt 
# numpy (math) libary
import numpy as np

from scipy import interpolate

path  = '../../../data/2017_12_19/'
```

```python
VOA = np.loadtxt(path+'1100-VOA_calibration.tsv')
VOA.shape
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)


ax1.plot(VOA[:,0],VOA[:,1])

plt.show()
plt.close()
```

```python
# normalize intensity in the range 0% - 100%
nVOA = VOA
nVOA[:,1] = ( nVOA[:,1]-min(nVOA[:,1]) )/( max(nVOA[:,1])-min(nVOA[:,1]) )
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)

ax1.plot( nVOA[:,1], nVOA[:,0])

plt.show()
plt.close()
```

```python
pf2 = np.polyfit(nVOA[:,1], nVOA[:,0], 2)
p2  = np.poly1d(pf2)

pf3 = np.polyfit(nVOA[:,1], nVOA[:,0], 3)
p3  = np.poly1d(pf3)

pf4 = np.polyfit(nVOA[:,1], nVOA[:,0], 4)
p4  = np.poly1d(pf4)

pf5 = np.polyfit(nVOA[:,1], nVOA[:,0], 5)
p5  = np.poly1d(pf5)

pf6 = np.polyfit(nVOA[:,1], nVOA[:,0], 6)
p6  = np.poly1d(pf6)

pf7 = np.polyfit(nVOA[:,1], nVOA[:,0], 7)
p7  = np.poly1d(pf7)
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)

xp = np.linspace(0, 1, 101)
ax1.plot(nVOA[::-1,1], nVOA[::-1,0], '.')
#ax1.plot(xp, p2(xp), '-')
ax1.plot(xp, p3(xp), '--')
#ax1.plot(xp, p4(xp), '--')
#ax1.plot(xp, p5(xp), '--')
#ax1.plot(xp, p6(xp), '--')
ax1.plot(xp, p7(xp), '--')

plt.show()
plt.close()
```

```python
tck = interpolate.splrep(nVOA[::-1,1], nVOA[::-1,0], s=0)
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)

xp = np.linspace(0, 1, 101)
ax1.plot(nVOA[:,1], nVOA[:,0], '.')
ax1.plot(xp, p7(xp), '--')
ynew = interpolate.splev(xp, tck, der=0)

plt.show()
plt.close()
```
