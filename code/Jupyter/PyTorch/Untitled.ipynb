```python
# plotting libraries
import matplotlib
import matplotlib.pyplot as plt 
# numpy (math) libary
import numpy as np

# torch library and sublibraries
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)

a,b,x0,c,x1 = [1.310, 41.516, 0.399, -42.491, 0.409]

x = np.linspace(-1,2,301)

y1 = x.copy()
y1[y1<0] = 0
y1 = a*y1
ax1.plot(x,y1)

y2 = x.copy()
y2[y2<x0] = 0
y2 = b*y2
#ax1.plot(x,y2)

y3 = x.copy()
y3[y3<x0] = 0
y3 = c*y3
#ax1.plot(x,y3)

y2b = x.copy()-x0
y2b[y2b<0] = 0
y2b = b*y2b
#ax1.plot(x,y2b)

y3b = x.copy()-x1
y3b[y3b<0] = 0
y3b = c*y3b
#ax1.plot(x,y3b)

ax1.plot(x,y1+y2b+y3b)

plt.show()
plt.close()
#y1 = x.clamp(min=0).mul_(p[0])
#y2 = x.clamp(min=p[2]).mul_(p[1])
#y3 = x.clamp(min=p[4]).mul_(p[3])
```
