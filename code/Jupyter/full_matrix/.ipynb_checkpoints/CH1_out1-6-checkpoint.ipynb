```python
# plotting libraries
import matplotlib
import matplotlib.pyplot as plt 
# numpy (math) libary
import numpy as np

# data dictionary
d = {}
```

```python
# Use LaTeX to manage text layout
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

```python
#################################################
######          FULL MATRIX CHIP           ######

folder = 'data/'
date1 = '2017-10-23'
date2 = '2017-10-24'
date3 = '2017-10-25'
date4 = '2017-10-27'
date5 = '2017-10-30'

path1 = folder+date1+'/'+date1
path2 = folder+date2+'/'+date2
path3 = folder+date3+'/'+date3
path4 = folder+date4+'/'+date4
path5 = folder+date5+'/'+date5
```

```python
#################################################
######       FIRST CHANNEL ANALYZED        ######

# Read CSV files:
# input 1, output 1
d['ch1o1'] = np.loadtxt(path1+'_P2mW_deg14_ch1o1_wlen1530-1550_medG.tsv')

# input 1, output 2
d['ch1o2'] = np.loadtxt(path1+'_P2mW_deg14_ch1o2_wlen1535-1550_medG.tsv')

# input 1, output 3
d['ch1o3'] = np.loadtxt(path1+'_P2mW_deg14_ch1o3_wlen1535-1550_medG.tsv')

# input 1, output 4
d['ch1o4'] = np.loadtxt(path1+'_P2mW_deg14_ch1o4_wlen1540-1555_medG.tsv')

# input 1, output 5
d['ch1o5'] = np.loadtxt(path1+'_P2mW_deg14_ch1o5_wlen1540-1560_medG.tsv')

# input 1, output 6
d['ch1o6'] = np.loadtxt(path1+'_P2mW_deg14_ch1o6_wlen1540-1560_medG.tsv')
```

```python
fig = plt.figure(figsize=(3*6.4, 2*4.8)) # default = 6.4, 4.8
ax1 = fig.add_subplot(111)
ax1.plot(d['ch1o1'][:,0], 10*np.log10(d['ch1o1'][:,2]), label='1st output')
ax1.plot(d['ch1o2'][:,0], 10*np.log10(d['ch1o2'][:,2]), label='2nd output')
ax1.plot(d['ch1o3'][:,0], 10*np.log10(d['ch1o3'][:,2]), label='3rd output')
ax1.plot(d['ch1o4'][:,0], 10*np.log10(d['ch1o4'][:,2]), label='4th output')
ax1.plot(d['ch1o5'][:,0], 10*np.log10(d['ch1o5'][:,2]), label='5th output')
ax1.plot(d['ch1o6'][:,0], 10*np.log10(d['ch1o6'][:,2]), label='6th output')

ax1.set_title(r'Full Matrix, first input channel, 1st-to-6th output channels', fontsize=16)
ax1.set_xlabel(r'Wavelength $\lambda$ $[nm]$', fontsize=16)
ax1.set_ylabel(r'Output intensity $[dBm]$', fontsize=16)

box = ax1.get_position()
ax1.set_position([box.x0, box.y0, box.width * 0.8, box.height])

legend = ax1.legend(loc='center left', fontsize=16, bbox_to_anchor=(1, 0.5))

plt.show()
plt.close()
```
